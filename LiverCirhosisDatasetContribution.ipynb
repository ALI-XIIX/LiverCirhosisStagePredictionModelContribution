```python
# ii. Univariate Analysis - Categorical Variables

# Create count plots for each categorical column
fig,axes = plt.subplots(nRows, nCols, figsize=(30,30))
for ind,col in enumerate(cat_columns):
    sns.countplot(y=col,data=df,ax=axes.flatten()[ind])
plt.show()


# iii. Bivariate Analysis - Categorical Variables by Stage

# Create count plots differentiated by Stage to understand relationships
print("Differentiation by stage")
fig,axes = plt.subplots(nRows, nCols, figsize=(25,25))
for ind,col in enumerate(cat_columns):
    sns.countplot(y=col,hue='Stage',data=df,ax=axes.flatten()[ind])
plt.show()


print("According to the differentiated graphs some generic conclusions can be made, that more females are prone to liver cirhosis, and the higher the stage of liver cirhosis the more patients have a status of death which follows the graph for status,also placebo seems to be the better drug as more stage 3 patients have been prescribed placebo, and finally out of the 4 nominal variables the symptom of spiders seems to be the most probable sign that someone has terrible liver cirhosis(stage 3)") 
print("")
```
